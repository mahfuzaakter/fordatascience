```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
#df = pd.DataFrame(np.random.randn(1000,4),columns=['a', 'b', 'c','d'])
%matplotlib inline
```

```python
# Check the head of the dataframe?
df.head()
```

```python
# You have created a dataset df. Plot c vs d as a scatter plot to recreate the plot below.
# Try to match the color and size of the data points. You should be able to match the size of the figure as well!
df.plot.scatter(x='c',y='d',color='green',figsize=(6,3))
```

```python
# ** Create a histogram of the 'd' column.**
df['d'].plot.hist(figsize=(4,3))
```

```python
# ** Change style sheets to set for 'bmh' and add df.plot.scatter(x='a',y='b',s=df['d']*60,color='g',figsize=(5,3))more bins (e.g. 50) to the above plot.**
plt.style.use('bmh')

df['d'].hist(bins=50,figsize=(5,3))
```

```python
# Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?
df.plot.scatter(x='a',y='b',s=df['d']*60,color='g',figsize=(5,3))
```

```python
# Recreate the plot below
df.plot.hexbin(x='a',y='b',gridsize=20,cmap='coolwarm',figsize=(4,3))
```

```python
# Create a boxplot comparing all columns in the data.
# Hint: use color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray') and sym = 'r+'

color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray') 
df.plot.box(color=color, sym = 'r+',figsize=(5,3))
```
