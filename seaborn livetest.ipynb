```python
import seaborn as sns
import matplotlib.pyplot as plt

data_dic = {
    'A': [12, 15, 45, 71],
    'B': [18, 26, 50, 80],
    'C': [32, 48, 78, 93],
    'D': [29, 35, 63, 81]
}


```

```python
# Create a DataFrame from the dictionary
import pandas as pd
df = pd.DataFrame(data_dic)

```

```python
# Create a heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(df, annot=True, cmap='rocket', linewidths=0.1, cbar=True)
plt
```
