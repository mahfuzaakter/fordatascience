```python

x = np.arange(0, 10)
y = x * 2
z = x ** 2

# Create a figure object
fig = plt.figure()

# Add an axis to the figure canvas
axes = fig.add_axes([0, 0, 1, 1])

# Plot the data x, y, and x, z on the axes with different colors and labels
axes.plot(x, y, label='y = 2x', color='blue')
axes.plot(x, z, label='z = x^2', color='green')
axes.plot(x, y, label='y = 2x (again)', color='red', linestyle='--')

# Set the labels for x and y
axes.set_xlabel('x')
axes.set_ylabel('y')

# Set the title
axes.set_title('Multiple Lines on the Plot')

# Create legends
axes.legend()

# Display the plot
plt.show()
```

```python


# Create a figure object and subplots
fig, axes = plt.subplots(3, 1,figsize=(5,3))
# Plot (x, y) on the first subplot
axes[0].plot(x, y, 'b--')


# Plot (x, z) on the second subplot
axes[1].plot(x, z, 'r-')

# Plot (x, y) on the third subplot
axes[2].plot(x, y,'g-' )


# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

# Step 1: Create a figure object
fig = plt.figure()

# Step 2: Add an axis to the figure canvas
axes = fig.add_axes([0, 0, 1, 1])

# Step 3: Plot the data x, y on the axes
axes.plot(x, y, label='y = 2x')

# Step 4: Set the labels for x and y
axes.set_xlabel('x')
axes.set_ylabel('y')

# Step 5: Set the title
axes.set_title('Simple Plot')

# Step 6: Create a legend
axes.legend()

# Display the plot
plt.show()
```

```python
# ** Create a figure object**
# ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
# ** Plot data (x,z) on main axes**
# ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
# ** Replicate the figure to axes, titles, legends etc**

x = np.arange(0,10)
y = x**2

fig=plt.figure(figsize=(6,3))
axes_main=fig.add_axes([0,0,1,1])
axes_insert=fig.add_axes([0.1,0.6,0.2,0.2])

axes_main.plot(x,y,'r-',label="y_main=x_main**2")
axes_main.set_xlabel('x_main')
axes_main.set_ylabel('y_main')
axes_main.set_title('title_main')


axes_insert.plot(x,y,'g-')
axes_insert.set_xlabel('x_insert')
axes_insert.set_ylabel('y_insert')
axes_insert.set_title('insert')

axes_insert.set_xlim([1,2])

axes_insert.set_ylim([0,5])
```
